```python
csv_file_path = '..\CSV\owid-covid-data.csv'
df = pd.read_csv(csv_file_path)
df.head()
```

```python
df["total_cases"]
```

```python
#make date into years, groupby location and year to sum the total cases. Then show only columns needed
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df_grouped = df.groupby(['location', 'year', 'iso_code']).agg({'new_cases': 'sum', 'population': 'first'}).reset_index()
df_grouped.rename(columns={'new_cases': 'Total Cases'}, inplace=True)

df_grouped[['iso_code','location', 'year', 'Total Cases', 'population']]
```

```python
df_grouped['Total Cases per 100,000'] = (df_grouped['Total Cases'] / df_grouped['population']) * 100000
df_grouped
```

```python
import pandas as pd
import geopandas as gpd

# Load the CSV file
csv_data = pd.read_csv('js/grouped_data1.csv')

# Load the GeoJSON file
geojson_data = gpd.read_file('js/countries.geojson')

```

```python
csv_data.head()
```

```python
geojson_data.head()
```

```python
# Select columns from the CSV file
csv_columns = ['name', 'alpha-2', 'alpha-3'] 
csv_selected = csv_data[csv_columns]

# Select columns from the GeoJSON file
geojson_columns = ['COUNTRY', 'ISO', 'AFF_ISO','geometry'] 
geojson_selected = geojson_data[geojson_columns]

```

```python
geojson_columns
```

```python
france_polygons2 = geojson_selected[geojson_selected['COUNTRY'] == 'United States']
france_polygons2
```

```python
france_polygons3 = csv_selected[csv_selected['name'] == 'United States']
france_polygons3
```

```python
import geopandas as gpd

# Merge the GeoJSON data with the CSV data based on AFF_ISO and alpha-2
merged_data = geojson_selected.merge(csv_selected, left_on='ISO', right_on='alpha-2')


# Create a GeoDataFrame from the merged data, assuming 'geometry' is the geometry column
geometry_column = 'geometry'  
merged_gdf=gpd.GeoDataFrame(merged_data, geometry=geometry_column)
merged_gdf = merged_gdf.rename(columns={'alpha-3': 'ISO_A3'})

# Save the GeoDataFrame as a GeoJSON file
output_file = 'merged.geojson'
merged_gdf.to_file(output_file, driver='GeoJSON')

```

```python
country_polygons = gpd.read_file('Madi/country_polygons.geojson')
merged_gdf = gpd.read_file('Madi/merged.geojson')

# Merge country_polygons and merged_gdf
merged_final = country_polygons.merge(merged_gdf, left_on='ISO_A3', right_on='alpha-3')

# drop the 'geometry_y' column
merged_final.drop(columns=['geometry_y'], inplace=True)

# fix'geometry_x'  to 'geometry'
merged_final = merged_final.rename(columns={'geometry_x': 'geometry'})

```
